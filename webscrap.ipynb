```python
from bs4 import BeautifulSoup
import requests

url = BeautifulSoup("http://www.nepalstock.com/floorsheet", 'html.parser')
response = requests.get(url)
print(response)
```

```python
data = response.text
soup = BeautifulSoup(data,'xml')
print(soup)
```

```python
print(type(soup))
```

```python
tags = soup.table
tags
```

```python
tag = tags.find_all('tr')
print(tag)
```

```python
y =[]
for i in tag:
  x = i.get_text()
  y.append(x)

print(y)
```

```python
final_data = y[1:-3]
final_data
```

```python
final_output = []
for i in final_data:
  list_data = i.split('\n')[1:-1]
  final_output.append(list_data)

final_output
```

```python
import csv
n = len(final_output)
f = open('sharemarket.csv','w')
writer = csv.writer(f)
for i in range(n):
  writer.writerow(final_output[i])
f.close()
```
