```python
from bs4 import BeautifulSoup
import requests
final_output = []
n = int(input("Enter n = "))
for d in range(1,n+1):
  url = BeautifulSoup(f"http://www.nepalstock.com/main/floorsheet/index/{d}/", 'html.parser')
  response = requests.get(url)
  data = response.text
  soup = BeautifulSoup(data,'xml')
  tags = soup.table
  tag = tags.find_all('tr')
  y =[]
  for i in tag:
    x = i.get_text()
    y.append(x)

  final_data = y[2:-3]
  
  for i in final_data:
    list_data = i.split('\n')[1:-1]
    final_output.append(list_data)
```

```python
final_output = [['S.N.','Contract No','Stock Symbol','Buyer Broke','Seller Broker','Quantity','Rate','Amount']]+final_output
```

```python
import csv
n = len(final_output)
f = open('sharemarket.csv','w')
writer = csv.writer(f)
for i in range(n):
  writer.writerow(final_output[i])
f.close()
```
